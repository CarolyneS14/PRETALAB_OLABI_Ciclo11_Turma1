# INTRODUÇÃO A BANCO DE DADOS

### Exemplo de Uso do SQLite com Python

1. Conectar ao banco de dados (ou criar um novo arquivo de banco de dados).

```python
import sqlite3
```

```python
 #Conectar (ou criar) um banco de dados chamado 'meu_banco.db'
conexao = sqlite3.connect('meu_banco.db')
```

```python
# Criar um cursor para executar comandos SQL
cursor = conexao.cursor()

print("Conectado ao banco de dados SQLite.")
```

2. Criar um cursor, que é o responsável por executar comandos SQL.

```python
# Excluir a tabela "Alunos"
cursor.execute("DROP TABLE IF EXISTS Alunos")

print("Tabela 'Alunos' excluída com sucesso.")
```

```python
# Criar uma tabela chamada "Alunos"
cursor.execute('''
CREATE TABLE IF NOT EXISTS Alunos (
    id INTEGER PRIMARY KEY,
    nome TEXT,
    idade INTEGER,
    disciplina TEXT,
    nota REAL
)
''')

print("Tabela 'Alunos' criada com sucesso.")
```

```python
# Consultar todos os dados da tabela "Alunos"
cursor.execute("SELECT * FROM Alunos")
resultados = cursor.fetchall()

# Exibir os dados recuperados
for linha in resultados:
    print(linha)
```

```python
# Consultar os nomes das colunas da tabela "Alunos"
cursor.execute("PRAGMA table_info(Alunos)")
colunas = cursor.fetchall()

# Exibir os nomes das colunas
for coluna in colunas:
    print(coluna[1])  # O nome da coluna está no índice 1
```

3. Executar comandos SQL, como criar tabelas, inserir, atualizar ou consultar dados.

```python
# Inserir registros na tabela "Alunos"
cursor.execute("INSERT INTO Alunos (nome, idade, disciplina, nota)  VALUES ('Maria', 15, 'Matemática', 9.5)"),
cursor.execute("INSERT INTO Alunos (nome, idade, disciplina, nota) VALUES ('João', 14, 'Matemática', 8.3)"),
cursor.execute("INSERT INTO Alunos (nome, idade, disciplina, nota) VALUES ('Jonas', 16, 'Matemática', 5.38)")

# Salvar as alterações no banco de dados
conexao.commit()

print("Dados inseridos com sucesso.")
```

```python
# Consultar todos os dados da tabela "Alunos"
cursor.execute("SELECT * FROM Alunos")
resultados = cursor.fetchall()

# Exibir os dados recuperados
for linha in resultados:
    print(linha)
```

```python
# Atualizar dados
cursor.execute("UPDATE Alunos SET NOTA = 9.0 WHERE NOME = 'Maria'")

conexao.commit()

print("Dados atualizados com sucesso.")
```

```python
# Deletar o aluno "Maria" da tabela
cursor.execute("DELETE FROM Alunos WHERE NOME = 'Maria'")

conexao.commit()

print("Dados deletados com sucesso.")
```

4. Fechar a conexão quando todas as operações forem concluídas.

```python
# Fechar a conexão com o banco de dados
conexao.close()

print("Conexão com o banco de dados fechada.")
```
