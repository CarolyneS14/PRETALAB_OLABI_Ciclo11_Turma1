##### Medidas de Tend√™ncia Central: Onde seus dados tendem a "morar" üè†

Essas medidas s√£o como tentar entender qual o "ponto de encontro" dos seus dados.

üëë**M√©dia: O l√≠der da equipe!**

A **m√©dia** √© aquele aluno que sempre "divide igual para todo mundo". Para calcular, voc√™ soma todos os valores e depois divide pelo n√∫mero de observa√ß√µes.

image.png







```python
# M√©dia
media = np.mean(dados)
media
```

‚úÇÔ∏è **Mediana: O mestre do equil√≠brio!**


A mediana √© o valor que est√° bem no meio dos dados ordenados. Se voc√™ colocar todos os n√∫meros em ordem, ela √© o valor que divide os dados em duas partes iguais.

Como calcular:

- Ordene os valores.

- Se o n√∫mero de observa√ß√µes for √≠mpar, a mediana √© o valor do meio.

- Se for par, a mediana √© a m√©dia dos dois valores do meio.

Exemplo:

Para as notas [6, 7, 8, 9, 10], a mediana √© 8, pois est√° no meio.

Se tivermos [6, 7, 8, 9], a mediana ser√°:

image.png


Mediana √© aquela pessoa justa que divide igualmente as opini√µes no grupo!

```python
# Mediana
mediana = np.median(dados)
mediana
```

üîÅ **Moda: A celebridade do grupo!**

A moda √© o valor que mais se repete nos dados. Se a m√©dia representa o "l√≠der" e a mediana o "justo", a moda √© a pessoa mais popular.

Exemplo:
Se temos as notas [6, 7, 7, 8, 9], a moda √© **7**, porque aparece duas vezes. Se n√£o houver repeti√ß√µes, dizemos que n√£o h√° moda.

```python
# Moda
from scipy import stats
moda = stats.mode(dados)
moda
```

A vari√¢ncia seria a m√©dia desses valores.

Se a vari√¢ncia √© grande, significa que os dados est√£o bem espalhados. Se √© pequena, est√£o mais "juntinhos".

```python
# Vari√¢ncia
variancia = np.var(dados)
variancia
```

üßÆ **Desvio Padr√£o: A dist√¢ncia t√≠pica da m√©dia **üö∂‚Äç‚ôÇÔ∏è

O desvio padr√£o √© a raiz quadrada da vari√¢ncia. Ele representa a "dist√¢ncia t√≠pica" entre cada valor e a m√©dia.

image.png

Exemplo: Se a vari√¢ncia das notas foi 2, ent√£o o desvio padr√£o ser√°:

image.png


O desvio padr√£o √© como medir o quanto as notas de um aluno "desviam" da m√©dia.

```python
# Desvio Padr√£o
desvio_padrao = np.std(dados)
desvio_padrao
```

```python
print(f"A vari√¢ncia √©: {variancia:.2f}, e o Desvio Padr√£o: {desvio_padrao:.2f}")
```

**Quartis e Percentis: Dividindo os dados em peda√ßos** üç∞

√Äs vezes, voc√™ quer dividir seus dados em partes iguais. Quartis e percentis ajudam a fazer isso.

Quartis: Dividem os dados em quatro partes iguais.

- O primeiro quartil (Q1) marca o ponto onde est√£o os 25% dos menores valores.

- O segundo quartil (Q2) √© a mediana (50%).

- O terceiro quartil (Q3) marca onde est√£o 75% dos menores valores.

- Percentis: Dividem os dados em 100 partes. O percentil 90 significa que 90% dos dados est√£o abaixo desse valor.


```python
# Quartil
primeiro_quartil = np.percentile(dados, 25)
segundo_quartil = np.percentile(dados, 50) #mediana
terceiro_quartil = np.percentile(dados, 75)

print(f"Primeiro quartil (25%): {primeiro_quartil}")
print(f"Segundo quartil (50%): {segundo_quartil}")
print(f"Terceiro quartil (75%): {terceiro_quartil}")
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Gerando o DataFrame fict√≠cio de vendas
np.random.seed(42)
```

```python
# Par√¢metros
n = 5000 # criando a quantidade de linhas do meu df
vendedores = [f"Vendedor {i}" for i in range (1, 21)] #quantidade de vendedores
produtos = [f"Produto {i}" for i in range (1, 11)] #quantidade de produtos
regioes = ['Sul', 'Sudeste', 'Norte', 'Centro-Oeste', 'Nordeste']
```

```python
# Gerando os Dados
data = pd.DataFrame({
    'vendedor': np.random.choice(vendedores, size=n),
    'produto': np.random.choice(produtos, size=n),
    'regiao': np.random.choice(regioes, size=n),
    'quantidade': np.random.randint(1, 50, size=n),
    'valor_unitario': np.round(np.random.uniform(10, 100, size=n), 2),
    'data_venda': pd.date_range('2023-01-01', periods=n, freq='h').strftime('%Y-%m-%d')
})
```

```python
#Visualizar df
data.head()
```

```python
# Visualizar colunas do df 'data'
data.info()
```

```python
# Corrigindo o coluna de 'data_venda'
data['data_venda'] = pd.to_datetime(data['data_venda'])
```

```python
# Visualizar colunas do df 'data'
data.info()
```

```python
# Calculando o valor total
data['valor_total'] = data['quantidade'] * data['valor_unitario']
```

```python
# Ver nova coluna
data.head()
```

- Estat√≠sticas descritivas

```python
# M√©dia
media_valor_total = round(data['valor_total'].mean())
media_valor_total
```

```python
# Mediana
mediana_valor_total = round(data['valor_total'].median())
mediana_valor_total
```

```python
# Moda (numero que mais aparece na coluna)
moda_valor_total = data['valor_total'].mode()[0]
moda_valor_total
```

```python
# Vari√¢ncia
variancia_valor_total = round(data['valor_total'].var(), 2)
variancia_valor_total
```

```python
# Desvio Padr√£o
desvio_padrao_valor_total = round(data['valor_total'].std(), 2)
desvio_padrao_valor_total
```

```python
# Quartil
quartis_valor_total = data['valor_total'].quantile([0.25, 0.5, 0.75])
quartis_valor_total
```

```python
# An√°lise descritiva da coluna
round(data['valor_total'].describe(), 2)
```

```python
# Filtro para ver o/os produto/s com o valor m√°ximo do total de venda na coluna 'valor_total'
data[data['valor_total'] == data['valor_total'].max()]
```

```python
# Filtro para verificar os produtos com o valor abaixo do primeiro quartil
data[data['valor_total'] < quartis_valor_total[0.25]]
```

```python
# Ou
# Filtro para verificar os produtos com o valor abaixo do primeiro quartil
data[data['valor_total'] <= 484.15]
```

```python
# Visualiza√ß√µes Histograma
plt.figure(figsize=(12, 6))
sns.histplot(data['valor_total'], kde=True, bins=30, color='blue')
plt.title('Distribui√ß√£o dos Valores Totais de Venda')
plt.xlabel('Valor Total de Venda')
plt.ylabel('Frequ√™ncia')
plt.show()
```

```python
# Boxplot dos valores totais de venda por regi√£o
plt.figure(figsize=(10, 6))
sns.boxplot(x='regiao', y='valor_total', data=data, palette='Set3', hue='regiao')
plt.title('Boxplot dos Valores Totais de Venda por Regi√£o')
plt.xlabel('Regi√£o')
plt.ylabel('Valor Total de Venda')
plt.show()
```

```python
# Gr√°fico de barras das quantidade vendidas por produtos
plt.figure(figsize=(12, 6))
sns.barplot(x='produto', y='quantidade', data=data, estimator=sum, palette='Set3', hue='produto')
plt.title('Total de Quantidade Vendida por Produto')
plt.xlabel('Produtos')
plt.ylabel('Quantidade Vendida')
plt.xticks(rotation=45)
plt.show()
```
