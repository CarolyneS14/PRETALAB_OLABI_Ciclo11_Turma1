#### Exemplo Prático de Probabilidade:

Vamos considerar um exemplo simples de lançamento de uma moeda. A probabilidade de cair "cara" ou "coroa" é 50%, ou seja, 0.5 para cada lado. Podemos simular isso em Python.

```python
import numpy as np
```

```python
# Simulação de 1000 lançamentos de uma moeda
np.random.seed(0)  # Para resultados reproduzíveis
lancamentos = np.random.choice(['Cara', 'Coroa'], size=1000, p=[0.5, 0.5])
```

```python
# Calculando a probabilidade de cada resultado
probabilidade_cara = np.sum(lancamentos == 'Cara') / len(lancamentos)
probabilidade_coroa = np.sum(lancamentos == 'Coroa') / len(lancamentos)
```

```python
print(f"Probabilidade de cair 'Cara': {probabilidade_cara:.2f}")
print(f"Probabilidade de cair 'Coroa': {probabilidade_coroa:.2f}")
```

Neste exemplo, simulamos 1000 lançamentos de uma moeda com probabilidades iguais de cair "cara" ou "coroa" (50%). Calculamos a frequência relativa de cada resultado para obter as probabilidades.

## Amostragem

Amostragem é como pegar uma "amostra grátis" no supermercado. Você experimenta uma parte para ter uma ideia de como é o produto todo. Em estatística, usamos a amostragem para entender as características de uma população sem precisar analisar todos os indivíduos.


#### Amostra Aleatória




- Amostra aleatória: É como jogar os nomes numa urna e sortear alguns. Cada nome (pessoa) tem a mesma chance de ser escolhido. É a forma mais básica e importante para garantir que sua amostra representa bem a população.


#### Tipos de Amostragem



##### Amostragem Simples

Na amostragem simples, todos os elementos da população têm a mesma probabilidade de serem escolhidos. É como tirar um nome de um chapéu: cada nome tem a mesma chance de ser selecionado. Aqui, a seleção é feita de forma completamente aleatória, sem levar em consideração nenhum fator específico.

1. **Como funciona?**

Imagine uma urna com 100 bolas numeradas de 1 a 100. Se quisermos selecionar 10 bolas, cada uma delas terá uma probabilidade igual (10%) de ser escolhida.

A seleção é feita aleatoriamente, sem considerar características específicas dos elementos.

***Vantagens:***

- Simplicidade: fácil de entender e executar.

- Tendência a representar bem a população quando ela é homogênea.

***Desvantagens:***

- Pode não ser eficaz se a população tiver grupos muito diferentes entre si (heterogênea).

- Pode não garantir que todos os grupos importantes sejam bem representados na amostra.


##### Amostragem Estratificada

A amostragem estratificada divide a população em subgrupos (estratos) com base em características específicas, como idade, gênero, localização, etc. Em seguida, são selecionadas amostras aleatórias de cada estrato proporcionalmente ao seu tamanho na população.

  **1.Como funciona?**

Imagine que queremos estudar a satisfação de clientes em uma rede de lojas que atende três regiões diferentes. Em vez de selecionar clientes aleatoriamente em toda a rede, dividimos a população em três estratos (regiões) e selecionamos amostras de cada região.

Se a Região A tem 50% dos clientes, a Região B tem 30% e a Região C tem 20%, a amostra será escolhida proporcionalmente (por exemplo, 50 pessoas da Região A, 30 da B e 20 da C, se o total da amostra for 100).

***Vantagens:***

- Garante que cada subgrupo importante da população seja representado na amostra.

- Aumenta a precisão das estimativas, especialmente quando os estratos são muito diferentes entre si.

***Desvantagens:***

- Pode ser mais complicado de implementar, pois requer conhecimento prévio das características da população.

- Se a definição dos estratos não for feita corretamente, a amostra pode não representar bem a população.

##### Aplicações na Análise de Dados


1. **Amostragem Simples:**

Boa para quando a população é homogênea, como, por exemplo, quando analisamos a idade média dos funcionários de uma mesma função em uma empresa.

2. **Amostragem Estratificada:**

Útil quando a população é diversificada e há necessidade de analisar características específicas, como diferentes faixas etárias em uma pesquisa sobre hábitos de consumo.

#### Exemplo de Amostragem

Vamos selecionar aleatoriamente uma amostra de clientes de uma base de dados fictícia para uma pesquisa de satisfação.

```python
import pandas as pd
```

```python
# Criando um DataFrame fictício com informações de clientes
dados_clientes = pd.DataFrame({
    'Cliente_ID': range(1, 1001),
    'Idade': np.random.randint(18, 70, size=1000),
    'Satisfacao': np.random.choice(['Alta', 'Média', 'Baixa'], size=1000, p=[0.5, 0.3, 0.2])
})
```

```python
# Visualizar
dados_clientes.head()
```

```python
# Amostra aleatória simples de 100 clientes
amostra_simples = dados_clientes.sample(n=100, random_state=0)
```

```python
print("Exemplo de Amostra Aleatória Simples:")
print(amostra_simples.head())
```

```python
# Verificar se realmente a amostra está com 100 clientes
print(amostra_simples.info())
```

```python
# Amostragem estratificada com base no nível de satisfação
amostra_estratificada = dados_clientes.groupby('Satisfacao', group_keys=False).apply(lambda x: x.sample(frac=0.1))
```

```python
print("\nExemplo de Amostragem Estratificada:")
print(amostra_estratificada.head())
```

```python
# Verificar a distribuição
print(amostra_estratificada['Satisfacao'].value_counts())
```

## Testes de Hipóteses

Os testes de hipóteses são como um tribunal: a hipótese nula é inocente até que se prove o contrário. A ideia é sempre começar assumindo que a hipótese nula é verdadeira e tentar encontrar evidências para rejeitá-la.



#### Hipótese Nula (H0) e Hipótese Alternativa (H1)

  


- Hipótese Nula (H0):

 É como dizer "não aconteceu nada de diferente".  É a afirmação de que não há efeito, diferença ou relação significativa nos dados que estamos analisando


 Por exemplo, você acha que uma nova dieta não faz diferença no peso das pessoas. Essa é a hipótese nula. É a posição conservadora, a suposição de que qualquer mudança observada é devida ao acaso.



  


- **Hipótese Alternativa (H1):**

É a "oposição" à hipótese nula. É a afirmação oposta à hipótese nula, sugerindo que existe um efeito, diferença ou relação significativa.


Aqui, você diz que a nova dieta faz diferença.
Esta é a posição que queremos provar, mostrando que há evidência suficiente nos dados para rejeitar a H0.

#### Testes Comuns

  


Vamos ver alguns dos testes de hipóteses mais usados e entender em que situações eles são aplicados:

##### Teste T (para uma ou duas amostras):






O teste T ajuda a verificar se essa diferença é significativa ou se pode ter ocorrido por acaso.

*Para uma amostra:*

Compara a média de uma amostra com uma média conhecida ou esperada.

Por exemplo, se o peso médio das pessoas em uma cidade é 70 kg, queremos saber se a média de peso das pessoas que seguiram a nova dieta é significativamente diferente de 70 kg.

*Para duas amostras:*

Compara as médias de dois grupos independentes.

Por exemplo, queremos saber se a altura média de meninos e meninas em uma escola é diferente.

***Exemplo Prático***

Imagine que você quer testar se a altura média dos alunos de duas turmas (Turma A e Turma B) é significativamente diferente. Vamos usar um teste T para verificar essa diferença.

```python
from scipy.stats import ttest_ind
```

```python
# Dados de altura (em cm) dos alunos das turmas
turma_a = np.random.normal(loc=170, scale=10, size=30)
turma_b = np.random.normal(loc=165, scale=12, size=30)
```

```python
# Aplicando o teste T
t_stat, p_value = ttest_ind(turma_a, turma_b)
```

```python
print(f"T-Estatística: {t_stat}")
print(f"Valor-p: {p_value}")
```

```python
if p_value < 0.05:
    print("Rejeitamos a hipótese nula: as médias das alturas são significativamente diferentes.")
else:
    print("Não rejeitamos a hipótese nula: não há diferença significativa nas médias das alturas.")
```

##### Teste Qui-Quadrado


Usado para testar se existe uma relação entre duas variáveis categóricas.

Esse teste verifica se a frequência observada de diferentes categorias difere da frequência esperada.

***Exemplo Prático:***

Imagine que queremos saber se a preferência por tipos de veículos (carro, moto, caminhão) varia de acordo com o gênero (masculino, feminino).

```python
from scipy.stats import chi2_contingency
```

```python
# Tabela de contingência
dados = [[50, 30, 20],  # Masculino: 50 carros, 30 motos, 20 caminhões
         [30, 20, 10]]  # Feminino: 30 carros, 20 motos, 10 caminhões
```

```python
# Aplicando o teste Qui-Quadrado
chi2_stat, p_value, dof, expected = chi2_contingency(dados)
```

```python
print(f"Estatística Qui-Quadrado: {chi2_stat}")
print(f"Valor-p: {p_value}")
```

```python
if p_value < 0.05:
    print("Rejeitamos a hipótese nula: há uma associação significativa entre o gênero e a preferência por veículos.")
else:
    print("Não rejeitamos a hipótese nula: não há associação significativa entre o gênero e a preferência por veículos.")
```

##### ANOVA (Análise de Variância)

O ANOVA é usado para comparar as médias de três ou mais grupos para verificar se há uma diferença significativa entre eles.  

***Exemplo Prático***

 Imagine que você quer comparar as notas médias de três turmas (Turma A, Turma B e Turma C) em uma prova de matemática.

```python
from scipy.stats import f_oneway
```

```python
# Notas das turmas
turma_a = np.random.normal(loc=70, scale=10, size=30)
turma_b = np.random.normal(loc=75, scale=8, size=30)
turma_c = np.random.normal(loc=80, scale=9, size=30)
```

```python
#Aplicando o ANOVA
f_stat, p_value = f_oneway(turma_a, turma_b, turma_c)
```

```python
print(f"F-Estatística: {f_stat}")
print(f"Valor-p: {p_value}")
```

```python
if p_value < 0.05:
    print("Rejeitamos a hipótese nula: há uma diferença significativa nas médias das notas entre as turmas.")
else:
    print("Não rejeitamos a hipótese nula: não há diferença significativa nas médias das notas entre as turmas.")
```

#### Valor-p



  valor-p é uma medida que nos ajuda a decidir se rejeitamos ou não a hipótese nula. Ele indica a probabilidade de observarmos os dados coletados (ou algo ainda mais extremo) se a hipótese nula for verdadeira.

- Valor-p baixo (geralmente menor que 0,05): Indica evidência forte contra a hipótese nula, sugerindo que devemos rejeitá-la. É como dizer: "Tem algo acontecendo aqui, não foi só sorte."


- Valor-p alto (maior que 0,05): Indica pouca evidência contra a hipótese nula, sugerindo que não devemos rejeitá-la. Nesse caso, os dados não fornecem suporte suficiente para rejeitar a hipótese nula.
